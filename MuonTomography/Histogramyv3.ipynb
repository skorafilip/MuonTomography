```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
x = []
y = []

with open("outputfile.dat", "r") as file:   
    for line in file:
        split = line.split('\t')
        x.append(float(split[0]))
        y.append(float(split[1]))
file.close()
signal = pd.DataFrame(
    {'x': x,
     'y': y,
    })
print(signal)

```

```python
plt.figure()
plt.hist2d(signal['x'],signal['y'],
          bins=500,  cmap=plt.cm.jet, range = [[0,10],[-10,20]],)
plt.savefig(fname = "histogram2d.png", dpi = 1000,)
plt.show()
```
